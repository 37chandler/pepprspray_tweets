```python
#useful imports for normalization and tokenization
import nltk
from collections import Counter
from nltk.corpus import stopwords
from pprint import pprint
from string import punctuation
import numpy as np
from nltk.probability import FreqDist
```

```python
sw = stopwords.words('english')
```

```python
#cleaning steps
pepper = open("20201019_Sentiment_tweets.txt",'r', encoding = 'UTF-8').read()
pepper_clean = [w for w in pepper.lower().split('\t')]
pepper_clean = [w for w in str(pepper_clean).split()]
pepper_clean = [w for w in pepper_clean if w.isalpha() and w not in sw]
#pepper_clean = [token for tokrn in pepper_clean]
```

```python
pepper_clean[:200]
```

```python
#tools to measure tken length and distribution
pepper_count = Counter(pepper_clean)
len(pepper_clean)
pepper_fd = FreqDist(pepper_clean)
pepper_fd
```

```python
#descriptive statistics tools
print(f"Sentiment is {len(pepper_clean)} tokens long.")
print(f"Sentiment has {len(set(pepper_clean))} unique tokens.")

print(f"Sentiment's lexical diversity is {len(set(pepper_clean))/len(pepper_clean):.3f}.")

# Build a vector of token length
pepper_token_len = [len(w) for w in pepper_clean]

print(f"Sentiment's average token length is {np.mean(pepper_token_len):.2f}.")

pprint(sorted(Counter(pepper_token_len).items()))

print("")
print("All statistics are calculated after normalization and tokenization.")
```
